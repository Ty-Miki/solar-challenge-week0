### Data cleaning

```python
# Drop comments column (Entirely empty)
day_time_df.drop(columns=["Comments"], inplace=True)
day_time_df.columns
```

Note -  The dataset contains 525,600 rows (which suggests 1 year of minute-level data: 365 days × 1440 mins)
- Considering this the data collected during nighttime is irrelevant for this EDA and should be dropped

```python
# Remove negative irradiance
irradiance_cols = ['GHI', 'DNI', 'DHI']
day_time_df = day_time_df[~day_time_df[irradiance_cols].lt(0).any(axis=1)]

# Drop nighttime rows: GHI ≤ 0
day_time_df = day_time_df[day_time_df['GHI'] > 0]
day_time_df.describe()
```

```python
import pandas as pd

# Convert Timestamp to datetime
day_time_df['Timestamp'] = pd.to_datetime(day_time_df['Timestamp'])
```

### Outlier Detection & Handling 

```python
from scipy.stats import zscore
import numpy as np

# Calculate z-scores for major columns
major_cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

for column in major_cols:
    z_scores = zscore(day_time_df[column])
    day_time_df[f'outlier_{column}'] = (np.abs(z_scores) > 3)

outlier_summary = {col: day_time_df[f'outlier_{col}'].sum() for col in major_cols}
print("Outlier counts per column:", outlier_summary)
```

##### Note

- No outliers for GHI, DNI, ModA and ModB
  - These variables have distributions that are well-behaved (or tightly clustered) during daytime.
  - No immediate action needed for outlier handling.


- DHI has 168 outliers
  - Worth visualizing — may include extreme low or high values.
  - Could be a result of sensor error or unusual atmospheric conditions.

- WS and WSgust have 600+ outliers:
  - Significant — may indicate: Very rare but real high wind events, Sensor noise, Data entry problems
  - Needs further investigation.

### TSA

```python
# Load cleaned data
cleaned_file_path = "../data/benin-malanville_clean.csv"
clean_df = load_csv(cleaned_file_path)
clean_df.head()
```

```python
# Aggregate relevant columns by hour
clean_df['Timestamp'] = pd.to_datetime(clean_df['Timestamp'])
clean_df['Hour'] = clean_df['Timestamp'].dt.hour
hourly_agg = clean_df.groupby('Hour')[['GHI', 'DNI', 'DHI', 'Tamb']].mean().reset_index()
hourly_agg.head()
```

```python
# Plot TSA for hourly data
plot_gen.plot_time_series(df=hourly_agg, columns=['GHI', 'DNI', 'DHI', 'Tamb'], time_column='Hour')
```

#### Note

- The hourly **TSA** reveals clear **diurnal patterns** in solar irradiance and temperature, with peak values **around midday**. Productive solar hours typically span **from late morning to early afternoon**, aligning with optimal PV generation periods. Temperature peaks lag behind irradiance, indicating potential efficiency impacts. These insights help in optimizing system performance and planning energy utilization strategies.

```python
# Aggregate relevant columns by day
clean_df['Date'] = clean_df['Timestamp'].dt.date
daily_agg = clean_df.groupby('Date')[['GHI', 'DNI', 'DHI', 'Tamb']].mean().reset_index()
daily_agg.head()
```

```python
# Plot TSA for daily data
plot_gen.plot_time_series(df=daily_agg, columns=['GHI', 'DNI', 'DHI', 'Tamb'], time_column='Date')
```

#### Note

- The daily **TSA** highlights seasonal trends in solar irradiance and temperature. GHI, DNI, and DHI show higher variability and peaks **during the dry season**, indicating stronger solar potential. Ambient temperature (Tamb) rises gradually, peaking **around mid-year**, consistent with seasonal climate patterns. These insights can support energy yield forecasting and maintenance planning.

```python
# Aggregate relevant columns by month
clean_df['Month'] = clean_df['Timestamp'].dt.month
monthly_agg = clean_df.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean().reset_index()
monthly_agg.head()
```

```python
# Plot TSA for monthly data
plot_gen.plot_time_series(df=monthly_agg, columns=['GHI', 'DNI', 'DHI', 'Tamb'], time_column='Month')
```

#### Note

- The monthly TSA reveals the following points
  - **GHI (Global Horizontal Irradiance)**: Peaks early in the year (Jan–Mar), dips mid-year (Jul–Aug), and rises again towards the end. This reflects seasonal sun angle and cloud cover.
  - **DNI (Direct Normal Irradiance)**: Shows a U-shaped pattern with lower values during the mid-year rainy season, indicating more cloud coverage and scattering.
  - **DHI (Diffuse Horizontal Irradiance)**: Peaks around March, suggesting increased atmospheric scattering (typical in humid or partially cloudy conditions), and dips during the dry months when the sky is clearer.
  - **Tamb (Ambient Temperature)**: Peaks between March and May, then decreases during the rainy season and slightly rises again later in the year—showing a typical tropical temperature pattern.

- These trends are useful for PV system planning, seasonal performance forecasting, and energy yield analysis, helping align maintenance schedules and optimize solar capture.

### Impact of cleaning

```python
# Group data set by cleaning
cleaning_groups = clean_df.groupby('Cleaning')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()
cleaning_groups.head()
```

#### Note

-  The groupd data for irradiance measurements is somewhat opposite from what is expected, normally dust blocks sunlight and uncleaned sensors would abasorb a lower amount of irradiance. this shows that sensors might be faulty.
  
- The grouped data for measurement is expected since dust causes overheating and could be the reason for higher temperature reading.

```python
# Draw box plots for grouped data
plot_gen.plot_box_grouped(clean_df, columns=["GHI", "DNI", "DHI", "Tamb"], group_column="Cleaning")
```

#### Note

- The box plot confirms that **irradiance sensors** that were not cleaned recorded higher values than cleaned sensors . **Further investigation into these sensor calibration or maintenance records is recommended.**

- The box plot also confirms **ambient temperature** readings are higher for uncleaned sensors than for cleaned sensors . **The result supports the impact of cleaning on temperature accuracy.**

```python
# Grouped TSA for furher analysis
plot_gen.plot_time_series_grouped(clean_df, columns=["GHI", "DNI", "DHI", "Tamb"], time_column="Timestamp", group_column="Cleaning")
```

#### Note

- Time series analysis reveals that irradiance and temperature measurements from uncleaned sensors often show higher and more erratic values compared to cleaned ones. This aligns with temperature expectations but contradicts expected irradiance behavior, suggesting potential sensor faults. The analysis also highlights the stabilizing effect of cleaning on sensor performance over time.



#### Conclusion on impact of cleaning

- Cleaning significantly improves sensor accuracy by reducing noise and erratic spikes in both irradiance and temperature measurements. Cleaned sensors provide more stable and realistic readings, especially for temperature, while discrepancies in irradiance highlight the importance of sensor maintenance beyond just cleaning.

#### Recomendation

- I recommend investigating sensor calibration and maintenance logs for those showing inconsistent or unexpectedly high irradiance values. Additionally, establish a more consistent and documented cleaning schedule to ensure reliable data quality and to facilitate clearer analysis of environmental impacts.
